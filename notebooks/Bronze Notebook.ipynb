```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
import requests
import json
from datetime import date , timedelta
```

```python
#Remove this before running Data Factory Pipeline
start_date = date.today() - timedelta(7) # 7 days
end_date = date.today() - timedelta(1)
```

```python
# API with start-date and end-date provided by data factora, formated by geojason output
url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={start_date}&endtime={end_date}"
```

```python
url
```

```python
try:
    # GET request to the API to fetch data.
    response = requests.get(url)

    #Check Response Status i.e if API returns an error like 404 or 500, it raises an exception
    response.raise_for_status()
    data = response.json().get('features', []) #convert the response into JSON and extract the ‘features’ field

    if not data:
     print("No data returned for the specified date range.")

    else:
    # specify fine name and path    
     file_path = f'/lakehouse/default/Files/{start_date}_earthquake_data.json'
    
    #save the json data
    with open(file_path, "w") as file:
     json.dump(data, file, indent=4)
    print(f"Data successfully saved to {file_path}")

except requests.exceptions.RequestException as e:
    print(f"Error fetching data from API: {e}")

```
